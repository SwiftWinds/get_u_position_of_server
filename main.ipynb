```python
def map_cabinet(file, cabinet):
    racks = pd.read_excel(file, sheet_name=cabinet)
    print("cabinet", cabinet)
    prev_u = None
    for index, row in racks.iterrows():
        if pd.isnull(row[1]) and prev_u is None:
            continue
        if not pd.isnull(row[1]):
            prev_u = row[1]
        if not pd.isnull(row[3]):
            if row[3] == "US02-5038ML-030-01":
                print("found US02-5038ML-030-01!")
            mapping[row[3]] = {"u_position": prev_u, "cabinet": cabinet}
```

```python
xl = pd.ExcelFile('US02-Cage1.xlsx')
```

```python
cabinets = xl.sheet_names[1:]
```

```python
for cabinet in cabinets:
    map_cabinet('US02-Cage1.xlsx', cabinet)
```

```python
for index, row in dfs.iterrows():
    if pd.isnull(row['所在U位']):
        print(row['标签'], row['机柜'], row['所在U位'])
        try:
            out = mapping[row['标签']]["u_position"]
            if out[-1] != "U":
                out += "U"
            dfs.at[index, '所在U位'] = out
        except KeyError:
            print(f"ERROR: {row['标签']} not found in mapping")
```

```python
def postprocess_u(u):
    if type(u) is datetime.datetime:
        return f"{u.strftime('%-m-%-d')}U"
    if type(u) is int or "U" not in u:
        return f"{u}U"
    if "\n" in u:
        return re.sub(r'^.*?\n', '', u)  # remove everything before and including a '\n'
    if "U" in u and u.index('U') != len(u) - 1:
        return re.sub(r'U(?=.*U)', "", u)  # replace all U's except last U
    return u
```

```python
del mapping["服务器各节点"]
```

```python
cleaned_mapping = {k: {"u_position": postprocess_u(v["u_position"]), "cabinet": v["cabinet"]} for k, v in
                   mapping.items()}
```

```python
with open("mapping.json", "w") as outfile:
    json.dump(cleaned_mapping, outfile, indent=4, sort_keys=True, default=str)
```
